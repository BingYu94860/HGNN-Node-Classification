```python
import json, os
import numpy as np
import matplotlib.pyplot as plt

folor_name = 'PGCN-CIFAR10' # PGCN-MNIST  PGCN-CIFAR10  CbPGCN-MNIST  CbPGCN-CIFAR10

root_list = [f"./{folor_name}/16/", f"./{folor_name}/32/", f"./{folor_name}/64/"] # , f"./{folor_name}/128/"
dict_dict_json = {}
for root in root_list:
    print(root)
    assert os.path.isdir(root)
    for file_name in os.listdir(root):
        with open(os.path.join(root, file_name), 'r') as f:
            dict_dict_json[file_name] = json.loads(f.read())
```

```python
num_theta, num_radius = 128, 32

history_accuracy_2d = []
history_val_accuracy_2d = []
history_loss_2d = []
history_val_loss_2d = []
for file_name, dict_value in dict_dict_json.items():
    if dict_value['num_theta'] == num_theta and dict_value['num_radius'] == num_radius:
        history_accuracy_2d.append(dict_value['history_accuracy'])
        history_val_accuracy_2d.append(dict_value['history_val_accuracy'])
        history_loss_2d.append(dict_value['history_loss'])
        history_val_loss_2d.append(dict_value['history_val_loss'])
```

```python
acc = np.mean(np.array(history_accuracy_2d), axis=0)
val_acc = np.mean(np.array(history_val_accuracy_2d), axis=0)
epoch_arr = np.arange(len(acc))+1

fig, ax = plt.subplots(figsize=(8, 4), dpi=300)
ax.set_xlim(left=0, right=len(acc) + 0.5)
ax.plot(epoch_arr, acc)
ax.plot(epoch_arr, val_acc)
# Chebyshev
ax.set_title('Polar Coordinate GCN Model Accuracy for CIFAR-10 dataset')
ax.set_ylabel('Accuracy')
ax.set_xlabel('Epoch')
ax.legend(['Train', 'Val'], loc='best')
#plt.close()
fig.savefig('Polar Coordinate GCN Model Accuracy for CIFAR-10 dataset.png', bbox_inches='tight', dpi=300, pad_inches=0.0)
```

```python
loss = np.mean(np.array(history_loss_2d), axis=0)
val_loss = np.mean(np.array(history_val_loss_2d), axis=0)
epoch_arr = np.arange(len(loss))+1

fig, ax = plt.subplots(figsize=(8, 4), dpi=300)
ax.set_xlim(left=0, right=len(loss) + 0.5)
ax.plot(epoch_arr, loss)
ax.plot(epoch_arr, val_loss)

ax.set_title('Polar Coordinate GCN Model Loss for CIFAR-10 dataset')
ax.set_ylabel('Loss')
ax.set_xlabel('Epoch')
ax.legend(['Train', 'Val'], loc='best')
#plt.close()
fig.savefig('Polar Coordinate GCN Model Loss for CIFAR-10 dataset.png', bbox_inches='tight', dpi=300, pad_inches=0.0)
```
