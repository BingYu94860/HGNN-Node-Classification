# Cora

```python
import os, json, collections
import matplotlib.pyplot as plt
import matplotlib as mpl

dataset_str = 'Cora'

path = 'output/GCN_cora_info~20210801_162724.json'
with open(path, 'r') as f:
    h_json = json.loads(f.read())
    
path = 'output/GCN_v2_cora_info~20210801_162807.json'
with open(path, 'r') as f:
    m_json = json.loads(f.read())
    
h_acc      = h_json['history_accuracy']
h_val_acc  = h_json['history_val_accuracy']
h_loss     = h_json['history_loss']
h_val_loss = h_json['history_val_loss']

m_acc      = m_json['history_accuracy']
m_val_acc  = m_json['history_val_accuracy']
m_loss     = m_json['history_loss']
m_val_loss = m_json['history_val_loss']
```

```python
# dataset_str = 'Cora'

fig, ax = plt.subplots(figsize=(8, 3), dpi=300)
ax.set_xlim(left=0, right=len(h_acc) * 1.01)
h_epoch = np.arange(len(h_acc)) + 1

ax.plot(h_epoch, h_acc, color='C0', label='GCN with dropout Train Acc', linestyle='-')
ax.plot(h_epoch, h_val_acc, color='C1', label='GCN with dropout Val Acc', linestyle='-')

ax.plot(h_epoch, m_acc, color='C2', label='GCN without dropout Train Acc', linestyle='-.')
ax.plot(h_epoch, m_val_acc, color='C3', label='GCN without dropout Val Acc', linestyle='-.')

ax.set_title(f'Accuracy curve of GCN Model in {dataset_str} dataset', fontsize=10)
ax.set_ylabel('Accuracy', fontsize=10)
ax.set_xlabel('Epoch', fontsize=10)
ax.tick_params(pad=0.5, labelsize=8)
ax.legend(loc='best')
#plt.close()
fig.savefig(f'Accuracy curve of GCN Model in {dataset_str} dataset.png', bbox_inches='tight', dpi=300, pad_inches=0.0)
```

```python
# dataset_str = 'Cora'

fig, ax = plt.subplots(figsize=(8, 3), dpi=300)
ax.set_xlim(left=0, right=len(h_acc) * 1.01)
h_epoch = np.arange(len(h_loss)) + 1

ax.plot(h_epoch, h_loss, color='C0', label='GCN with dropout Train Loss', linestyle='-')
ax.plot(h_epoch, h_val_loss, color='C1', label='GCN with dropout Val Loss', linestyle='-')

ax.plot(h_epoch, m_loss, color='C2', label='GCN without dropout Train Loss', linestyle='-.')
ax.plot(h_epoch, m_val_loss, color='C3', label='GCN without dropout Val Loss', linestyle='-.')

ax.set_title(f'Loss curve of GCN Model in {dataset_str} dataset', fontsize=10)
ax.set_ylabel('Loss', fontsize=10)
ax.set_xlabel('Epoch', fontsize=10)
ax.tick_params(pad=0.5, labelsize=8)
ax.legend(loc='best')
#plt.close()
fig.savefig(f'Loss curve of GCN Model in {dataset_str} dataset.png', bbox_inches='tight', dpi=300, pad_inches=0.0)
```
